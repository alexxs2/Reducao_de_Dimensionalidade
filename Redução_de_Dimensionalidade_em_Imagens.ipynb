```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('lena.jpg')
fig = plt.figure(figsize=(8, 8))
plt.imshow(img)
plt.show()
```

```python
width, height = img.size
# Transformando a imagem em tons de cinza
gray_img = Image.new('L', (width, height))
for x in range(width):
    for y in range(height):
        r, g, b = img.getpixel((x, y))
        # fórmula de luminância
        gray = int(0.2989 * r + 0.5870 * g + 0.1140 * b)
        gray_img.putpixel((x, y), gray)
```

```python
# Transformando a imagem binarizada a partir do cinza
threshold = 128
bin_img = Image.new('L', (width, height))
for x in range(width):
    for y in range(height):
        g = gray_img.getpixel((x, y))
        bin_value = 255 if g > threshold else 0
        bin_img.putpixel((x, y), bin_value)
```

# **Visualização das transformações**

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(img)
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(gray_img, cmap='gray')
axes[1].set_title('Tons de Cinza')
axes[1].axis('off')

axes[2].imshow(bin_img, cmap='gray')
axes[2].set_title('Binarizada')
axes[2].axis('off')

plt.show()
```
